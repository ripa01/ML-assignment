```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Heart Disease.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# Data preprocessing


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
x=df.drop(columns=['HeartDisease'])
y=df['HeartDisease']
```

# Undersampling

```python
from imblearn.under_sampling import NearMiss
```

```python
NearMiss_obj = NearMiss()
new_x , new_y = NearMiss_obj.fit_resample(x,y)
```

```python
new_y.value_counts()
```

```python
df2=pd.DataFrame(new_x)
df2.head()
```

```python
df3=pd.DataFrame(new_y)
df3.head()
```

```python
df4=pd.concat([df2,df3],axis=1)
df4
```

```python
df4.nunique()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
mms = MinMaxScaler() 
```

```python
scale_x = mms.fit_transform(new_x)
```

```python
pca = PCA(n_components=3)
x_pca = pca.fit_transform(scale_x)
```

```python
x_pca
```

# Split Dataset into Train and Test


```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import r2_score,mean_absolute_error,accuracy_score,mean_squared_error
```

```python
xtrain,xtest,ytrain,ytest=tts(new_x,new_y,train_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
def model(model):
    model.fit(xtrain, ytrain)
    y_pred = model.predict(xtest)
    r2 = r2_score(ytest, y_pred)
    mse = mean_squared_error(ytest, y_pred)
    mae = mean_absolute_error(ytest, y_pred)
    acc = model.score(xtest, ytest)
    print(str(model)+ ' --> ' +str(acc))
    print(f"R2 Score: {r2:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"MAE: {mae:.4f}")
    
```

```python
lr = LogisticRegression()
model(lr)

dt = DecisionTreeClassifier()
model(dt)

rf = RandomForestClassifier()
model(rf)

knn =KNeighborsClassifier(n_neighbors= 50)
model(knn)

svm = SVC(kernel='linear')
model(svm)
```

```python
from sklearn.metrics import classification_report
```

```python
pred_lr=lr.predict(xtest)
pred_dt = dt.predict(xtest)
pred_knn = knn.predict(xtest)
pred_svm=svm.predict(xtest)
pred_rf = rf.predict(xtest)
```

```python
print("Report of Logistic Regression:\n\n",classification_report(ytest, pred_lr))
print("Report of Knearest Neighbour Classificaiton:\n\n",classification_report(ytest, pred_knn))
print("Report of Support Vector Machine Classification:\n\n",classification_report(ytest, pred_svm))
print("Report of Random Forest Classification:\n\n",classification_report(ytest, pred_rf))
print("Report of Decision Tree:\n\n",classification_report(ytest, pred_dt))
```
